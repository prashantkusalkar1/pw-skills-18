Question 2 : Explain the role of the following Tableau components:
a) Data Pane
b) Worksheet
c) Dashboard
d) Story

a) Data Pane
Located on the left side of the worksheet
Displays all available data fields from your dataset
Contains:
Dimensions (categorical data: Name, Region, Category)
Measures (numerical data: Sales, Profit)

b) Worksheet
A single workspace/tab where you create visualizations
 Role:
Used to build:
Charts (bar, line, pie, etc.)
Graphs and tables
Each worksheet = one visualization

c) Dashboard
A collection of multiple worksheets displayed on one screen
 Role:
Combines different visuals into a single interactive view
Allows:
Filters across multiple charts
Better data storytelling

d) Story
A sequence of dashboards or worksheets presented step-by-step
Role:
Used for data storytelling and presentations
Helps explain insights in a logical flow

Question 3 : What is the difference between Dimensions and Measures in Tableau?
Provide examples of each.

Dimensions vs Measures in Tableau
🔹 Dimensions
Qualitative (categorical) data
Used to describe or group data
Usually text, dates, or IDs
Not aggregated by default
Role:
Break data into categories for analysis

Measures
Quantitative (numerical) data
Used for calculations and analysis
Automatically aggregated (SUM, AVG, COUNT, etc.)
Role:
Provide values that can be measured and analyzed

Question 4 : Define and explain the purpose of Filters, Parameters, and Sets in
Tableau.


1. Filters
 Definition:

Filters are used to limit the data displayed in a visualization by applying conditions.

 Purpose:
Show only relevant data
Remove unwanted data
Focus analysis on specific values

2. Parameters
Definition:

Parameters are user-defined dynamic values that allow users to control calculations, filters, or visuals.

 Purpose:
Provide flexibility and interactivity
Allow users to change values manually

3. Sets
Definition:

Sets are custom groups of data based on conditions or selection.

 Purpose:
Segment data into meaningful groups
Compare selected vs remaining data

Question 5 : Create a bar chart showing Gross Sales by Country.
● 🔗 Dataset Link:Global_sales_dataset
● Sort the countries in descending order of sales
● Highlight or annotate the bar that represents the maximum and minimum
Gross Sales.
● Add data labels and format the chart for presentation.

Step 1: Load Dataset
Open Tableau Desktop
Connect to Global_sales_dataset
Go to a new Worksheet

Step 2: Create Bar Chart
 Drag Fields:
Country → Columns
Gross Sales → Rows

Step 3: Sort in Descending Order
Click on the Sort icon (on toolbar)
OR
Right-click Country → Sort → Descending by Gross Sales

Step 4: Add Data Labels
Click on Marks Card → Label
Check Show Mark Labels

Step 5: Highlight Max & Min Sales
✅ Method 1: Using Calculated Field

Create a calculated field:

IF SUM([Gross Sales]) = WINDOW_MAX(SUM([Gross Sales])) THEN "Max"
ELSEIF SUM([Gross Sales]) = WINDOW_MIN(SUM([Gross Sales])) THEN "Min"
ELSE "Others"
END
